```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.dtypes
```

 Make a plot of total bill vs tip and fit a line of best fit to it using the tips dataset

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue = 'smoker', scatter = False)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'smoker', size='size', legend = False)
```

```python
sns.catplot(x="smoker", y="tip", hue="day", data=tips, kind="violin")
```

```python
agg_data = tips.pivot_table(index="sex", 
               columns='day', 
               values='tip',
               aggfunc='mean')

sns.heatmap(agg_data)
```

```python
pivoted = tips.pivot_table(index="time", columns='size', values='total_bill', aggfunc='mean')
sns.heatmap(pivoted)
```
